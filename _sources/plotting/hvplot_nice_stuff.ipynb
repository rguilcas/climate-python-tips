# Introducing hvplot
Hvplot is a super intuitive, really nice plotting tool that allows you to create interactive plots, from bar charts, to maps and line plots. I will here demonstrate som of its capabilities.
## Estimated inversion Strenght in ERA5

```python
import xarray as xr 
import hvplot.xarray
import geoviews.feature as gf
import geoviews as gv
import cartopy.crs as ccrs
from geoviews import opts
import cartopy.feature as cf
import panel as pn
import panel.widgets as pnw
import holoviews as hv
```

```python
hv.output(widget_location='bottom')
data_path = '/home/robin/Climserv/data/Observations/ERA20C/EIS_ERA20C_freq-year.nc'
land_mask_path = '/home/robin/Climserv/data/Observations/ERA5/lsm/sftlf_global_ERA20C.nc'

data = xr.open_dataset(data_path, chunks = dict(time=12))
data['time'] = (data.time).astype(int)
land = xr.open_dataset(land_mask_path).sftlf

data_masked = data.where(land<1)

proj = ccrs.Robinson(central_longitude=210)
eis = data_masked.hvplot(x='lon', y='lat', 
                         geo=True, project=True, projection=proj, 
                         cmap='magma_r', coastline=True,clim=(-2,15),
                         widget_type='scrubber', widget_location='bottom')

land_feat = gf.land().opts(opts.Feature(fill_color='lightgrey'))
eis
#graticules = cf.NaturalEarthFeature(
#    category='physical',
#    name='graticules_30',
#    scale='110m')
#eis*gv.Feature(graticules, group='Lines').opts(opts.Feature('Lines',color='lightgrey'))*land_feat*gf.coastline
```
